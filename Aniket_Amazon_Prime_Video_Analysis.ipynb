## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Create a copy of datasets for analysis
titles_clean = titles_df.copy()
credits_clean = credits_df.copy()

# 1. Handle missing values in seasons column (fill 0 for movies)
titles_clean['seasons'] = titles_clean['seasons'].fillna(0)

# 2. Create a numeric age certification mapping for analysis
age_mapping = {
    'G': 1, 'PG': 2, 'PG-13': 3, 'R': 4, 'NC-17': 5,
    'TV-Y': 1, 'TV-Y7': 2, 'TV-G': 2, 'TV-PG': 3, 'TV-14': 4, 'TV-MA': 5
}
titles_clean['age_rating_numeric'] = titles_clean['age_certification'].map(age_mapping)

# 3. Extract first genre for analysis
titles_clean['primary_genre'] = titles_clean['genres'].apply(
    lambda x: x.split(',')[0] if pd.notna(x) else None
)

# 4. Extract first production country for analysis
titles_clean['primary_country'] = titles_clean['production_countries'].apply(
    lambda x: x.split(',')[0] if pd.notna(x) else None
)

# 5. Create decade column from release_year
titles_clean['decade'] = (titles_clean['release_year'] // 10) * 10

# 6. Create content quality categories
def quality_category(score):
    if pd.isna(score):
        return 'Unknown'
    elif score >= 8.0:
        return 'Excellent'
    elif score >= 7.0:
        return 'Good'
    elif score >= 6.0:
        return 'Average'
    else:
        return 'Below Average'

titles_clean['quality_category'] = titles_clean['imdb_score'].apply(quality_category)

# 7. Merge credits data to get director and actor information
# Get directors for each title
directors = credits_clean[credits_clean['role'] == 'DIRECTOR'][['id', 'name']].drop_duplicates()
directors = directors.groupby('id')['name'].apply(lambda x: ', '.join(x)).reset_index()
directors.columns = ['id', 'director']

# Get actor count for each title
actor_counts = credits_clean[credits_clean['role'] == 'ACTOR'].groupby('id').size().reset_index()
actor_counts.columns = ['id', 'actor_count']

# Merge with titles
titles_clean = titles_clean.merge(directors, on='id', how='left')
titles_clean = titles_clean.merge(actor_counts, on='id', how='left')
titles_clean['actor_count'] = titles_clean['actor_count'].fillna(0).astype(int)

# 8. Create has_director flag
titles_clean['has_director'] = titles_clean['director'].notna()

# Display the cleaned dataset information
print("Cleaned Dataset Info:")
display(titles_clean.info())

print("\nCleaned Dataset First 5 rows:")
display(titles_clean.head())
````

### What all manipulations have you done and insights you found?

Data Manipulations Performed:

Handled Missing Values: Filled missing 'seasons' values with 0 (as they represent movies)

Created Numeric Age Rating: Mapped age certifications to numeric values for analysis

Extracted Primary Genre: Created a column with the first genre from the list

Extracted Primary Country: Created a column with the first production country

Created Decade Column: Grouped release years into decades for trend analysis

Created Quality Categories: Categorized content based on IMDb scores

Merged with Credits Data: Added director information and actor counts to titles

Created Has Director Flag: Indicator for whether a title has director information

Key Insights from Wrangling:
